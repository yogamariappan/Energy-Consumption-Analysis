```python

print(data.isnull().sum().sort_values(ascending=False))
```

```python
data.dropna(subset=['Address'], inplace=True)
data.dropna(subset=['Postal Code'], inplace=True)
```

```python
data.describe()
```

```python
Q1 = data['Property GFA - Self-Reported (m)'].quantile(0.25)
Q3 = data['Property GFA - Self-Reported (m)'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)


print(lower_bound ,upper_bound)
```

```python
# Check for outliers
outliers_present = ((data['Property GFA - Self-Reported (m)'] < lower_bound) | (data['Property GFA - Self-Reported (m)'] > upper_bound)).any()

if outliers_present:
    print("Outliers are present.")
else:
    print("No outliers are present.")
```

```python
data['Property GFA - Self-Reported (m)'] = data['Property GFA - Self-Reported (m)'].clip(lower=lower_bound, upper=upper_bound)
```

```python
Q1 = data['Number of Buildings'].quantile(0.25)
Q3 = data['Number of Buildings'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)


print(lower_bound ,upper_bound)
```

```python
# Check for outliers
outliers_present = ((data['Number of Buildings'] < lower_bound) | (data['Number of Buildings'] > upper_bound)).any()

if outliers_present:
    print("Outliers are present.")
else:
    print("No outliers are present.")
```
